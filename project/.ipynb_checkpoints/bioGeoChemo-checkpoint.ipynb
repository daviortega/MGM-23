```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import rcParams
rcParams.update({'figure.autolayout': False})

rc={'lines.linewidth': 2, 'axes.labelsize': 18, 'axes.titlesize': 18}
sns.set(rc=rc)

# The following is specific Jupyter notebooks
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}

import bokeh.io
import bokeh.mpl
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
chemoDataCounts = pd.read_csv('functionCounts.csv')
metaData = pd.read_csv('metadataRefMetagenomes.csv')
metaData.columns.values
```

```python
metaData['Both'] = metaData['Gene Count   * assembled'] + metaData['Gene Count   * unassembled']
```

```python
chemoDataCountsDF = pd.melt(chemoDataCounts, id_vars=['Function ID', 'Name'])
 #= chemoDataCountsDF['CheA'] / metaData['Gene Count   * assembled']
chemoDataCountsDF
```

```python
chemoDataCountsDF['Che Normalized (* 1E6)'] = 0
chemoDataCountsDF['Habitat'] = ''
chemoDataCountsDF['GOLD Sequencing Depth'] = ''

for i, data in chemoDataCountsDF.iterrows():
    chemoDataCountsDF['Che Normalized (* 1E6)'][i] = 1000000 * chemoDataCountsDF['value'][i] / metaData[metaData['taxon_oid'] == int(data['variable'].split(' ')[3])]['Gene Count   * assembled'].iloc[0]
    chemoDataCountsDF['Habitat'][i] = metaData[metaData['taxon_oid'] == int(data['variable'].split(' ')[3])]['Habitat'].iloc[0]
   #chemoDataCountsDF['GOLD Sequencing Depth'][i] = metaData[metaData['taxon_oid'] == int(data['variable'].split(' ')[3])]['GOLD Sequencing Depth'].iloc[0]
```

```python
chemoDataCountsDF
```

```python
for i in set(chemoDataCountsDF['Name']):
    g = sns.swarmplot(data=chemoDataCountsDF[chemoDataCountsDF['Name'] == i], y='Habitat', x='Che Normalized (* 1E6)', hue='Habitat')
    for label in g.get_xticklabels():
        label.set_rotation(90)
    g.legend_.remove()
    plt.title(i)
    plt.subplots_adjust(top=0.88)
    #plt.gcf().subplots_adjust(bottom=0.15)
    #print(chemoDataCountsDF[chemoDataCountsDF['Name'] == i]['Function ID'])
    filename = chemoDataCountsDF[chemoDataCountsDF['Name'] == i]['Function ID'].iloc[0]
    plt.savefig(filename + '.pdf')
    plt.show()
```

```python
g = sns.FacetGrid(chemoDataCountsDF, col="Function ID", col_wrap=4)
g.map(sns.swarmplot, data=chemoDataCountsDF, x='Habitat', y='Che Normalized (* 1000000)')

#for label in g.get_xticklabels():
#label.set_rotation(90)
    
#g.legend_.remove()
```
